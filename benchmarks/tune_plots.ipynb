```python
import pandas as pd
import glob
import re
import plotly.graph_objects as go
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
from scipy import stats

import plotly.io as pio
pio.renderers.default = "notebook"
```

```python
directory_path = "results_nonlocal_all"
file_list = glob.glob("**/*.parquet", root_dir=directory_path, recursive=True)
file_list = [file for file in file_list if "tmp" not in file]
print(f"Found {len(file_list)} files")
```

```python
plt.rcParams["font.family"] = "serif"
plt.rcParams["mathtext.fontset"] = "dejavuserif"
plt.rcParams['pdf.fonttype'] = 42
plt.rcParams['ps.fonttype'] = 42
plt.rcParams['font.size'] = 18
```

```python
# Configuration
cur_metric = 'ate'  # or 'spill'
save = True
use_log_scale = True  # Use log scale for x-axis
use_confidence_interval = True  # If True, use 95% CI; if False, use std dev
confidence_level = 0.95  # Confidence level (0.95 for 95% CI)

# Define parameters and their display names
params = ['config/beta_max', 'config/encoder_conv2', 'config/weight_decay_cvae', 'config/lr_cvae']
param_names = [r'$\beta$', '$d_Z$', 'Weight Decay', 'Learning Rate']

# Group files by dataset and dataset_index
dataset_groups = {}
for file in file_list:
    folders = file.split('/')
    dataset_name = folders[0]
    if dataset_name not in dataset_groups:
        dataset_groups[dataset_name] = []
    dataset_groups[dataset_name].append(file)

# Process each dataset
for dataset_name, dataset_files in dataset_groups.items():
    print(f"\nProcessing dataset: {dataset_name}")
    
    # Process each dataset_index (0 and 1)
    for dataset_idx in range(2):
        print(f"  Dataset index: {dataset_idx}")
        
        # Collect data for all radii
        all_data = []
        
        for file in dataset_files:
            df = pd.read_parquet(directory_path + "/" + file)
            df = df[df['ate'].notna()]
            df = df[df['dataset_index'] == dataset_idx]
            if len(df) == 0:
                print(file)
                continue
            
            folders = file.split('/')
            s = folders[1]
            match = re.search(r'_r(\d+)_(\w+)', s)
            if match:
                radius = int(match.group(1))
                model = match.group(2).replace('_tune', '')
            
            df['radius'] = radius
            df['model'] = model
            df['config/encoder_conv2'] = df['config/encoder_conv2'] // 2
                        
            all_data.append(df)
                            
        # Combine all data
        combined_df = pd.concat(all_data, ignore_index=True)
        
        # colors = plt.cm.viridis(np.linspace(0, 1, len(combined_df['radius'].unique())))
        # radius_color_map = dict(zip(sorted(combined_df['radius'].unique()), colors))
        default_colors = plt.rcParams['axes.prop_cycle'].by_key()['color']
        radii = sorted(combined_df['radius'].unique())
        radius_color_map = {radius: default_colors[i % len(default_colors)] for i, radius in enumerate(radii)}

        # Create separate figure for each parameter
        for idx, (param, param_name) in enumerate(zip(params, param_names)):
            fig, ax = plt.subplots(figsize=(8, 6))
            
            # For each radius, compute mean and std of metric for each parameter value
            for radius in sorted(combined_df['radius'].unique()):
                radius_df = combined_df[combined_df['radius'] == radius]
                
                # Remove outliers for this radius
                Q1 = radius_df[cur_metric].quantile(0.25)
                Q3 = radius_df[cur_metric].quantile(0.75)
                IQR = Q3 - Q1
                radius_df = radius_df[
                    (radius_df[cur_metric] >= Q1 - 1.5 * IQR) & 
                    (radius_df[cur_metric] <= Q3 + 1.5 * IQR)
                ]
                
                if radius == 0 and cur_metric == 'spill':
                    continue
                
                # Determine if we need to bin (for continuous parameters)
                unique_vals = radius_df[param].nunique()
                
                if unique_vals > 10:  # If more than 10 unique values, bin them
                    # Create bins
                    n_bins = 8
                    if use_log_scale:
                        # Create log-spaced bins
                        param_min = radius_df[param].min()
                        param_max = radius_df[param].max()
                        if param == 'config/encoder_conv2':
                            bins = np.logspace(np.log2(param_min), np.log2(param_max), n_bins + 1, base=2)
                        else:
                            bins = np.logspace(np.log10(param_min), np.log10(param_max), n_bins + 1)
                    else:
                        bins = n_bins
                    
                    # Bin the data
                    radius_df['param_bin'] = pd.cut(radius_df[param], bins=bins)
                    
                    # Compute bin centers for plotting
                    param_values = []
                    means = []
                    errors = []
                    
                    for bin_interval in radius_df['param_bin'].cat.categories:
                        subset = radius_df[radius_df['param_bin'] == bin_interval][cur_metric]
                        if len(subset) > 0:
                            # Use bin center
                            bin_center = (bin_interval.left + bin_interval.right) / 2
                            param_values.append(bin_center)
                            means.append(subset.mean())
                            
                            # Calculate error (std or CI)
                            if use_confidence_interval:
                                # 95% confidence interval
                                n = len(subset)
                                std_err = subset.std() / np.sqrt(n)
                                # Use t-distribution for small samples
                                t_val = stats.t.ppf((1 + confidence_level) / 2, n - 1)
                                error = t_val * std_err
                            else:
                                # Standard deviation
                                error = subset.std()
                            
                            errors.append(error)
                else:
                    # Use discrete parameter values directly
                    param_values = sorted(radius_df[param].unique())
                    means = []
                    errors = []
                    
                    for pval in param_values:
                        subset = radius_df[radius_df[param] == pval][cur_metric]
                        means.append(subset.mean())
                        
                        # Calculate error (std or CI)
                        if use_confidence_interval:
                            # 95% confidence interval
                            n = len(subset)
                            std_err = subset.std() / np.sqrt(n)
                            # Use t-distribution for small samples
                            t_val = stats.t.ppf((1 + confidence_level) / 2, n - 1)
                            error = t_val * std_err
                        else:
                            # Standard deviation
                            error = subset.std()
                        
                        errors.append(error)
                
                # Convert to numpy arrays
                param_values = np.array(param_values)
                means = np.array(means)
                errors = np.array(errors)
                
                # Plot with error bars
                color = radius_color_map[radius]
                error_label = f'{int(confidence_level*100)}% CI' if use_confidence_interval else 'Std Dev'
                ax.errorbar(param_values, means, yerr=errors, 
                           label=f'r={radius}', marker='o', capsize=5,
                           color=color, linewidth=2, markersize=8, alpha=0.7)
            
            # Set labels and title
            if use_log_scale:
                ax.set_xscale('log')
                if param == 'config/encoder_conv2':
                    ax.set_xlabel(f'$\log_2$({param_name})', fontsize=18)
                else:
                    ax.set_xlabel(f'$\log_{{10}}$(' + param_name + ')', fontsize=18)
            else:
                ax.set_xlabel(param_name, fontsize=18)
            
            if cur_metric == 'ate':
                ax.set_ylabel('DIR', fontsize=18)
            else:
                ax.set_ylabel(cur_metric.upper(), fontsize=18)
            ax.grid(False, alpha=0.0)
            ax.legend(loc='best')
            
            plt.tight_layout()
            
            # Save individual figure
            if save:
                output_dir = Path("plots")
                output_dir.mkdir(exist_ok=True)
                error_suffix = "ci" if use_confidence_interval else "std"
                # Clean parameter name for filename
                param_clean = param.replace('config/', '').replace('/', '_')
                fname = f"{dataset_name}_{param_clean}_{cur_metric}_idx{dataset_idx}_{error_suffix}"
                if use_log_scale:
                    fname += "_log"
                plt.savefig(output_dir / f"{fname}.pdf", bbox_inches='tight', dpi=300)
                print(f"    Saved: plots/{fname}.pdf")
            
            plt.show()
            plt.close()
```

```python
import os
from pathlib import Path

# Configuration
plots_dir = "plots"  # Directory where PDFs are saved
output_file = "sensitivity_figures2.tex"

# Mapping for dataset names to display names and their unobserved confounders
dataset_info = {
    'small_healthdgrid': {
        'display_name': 'PM_{2.5} \\;\\to\\; m',
        'idx0_confounder': 'q_{\\text{summer}}',
        'idx1_confounder': '\\rho_{\\text{pop}}',
    },
    'pm25grid': {
        'display_name': 'SO_4 \\;\\to\\; PM_{2.5}',
        'idx0_confounder': 'OC',
        'idx1_confounder': 'NH_{4}'
    }
}

# Mapping for rd values in filenames
rd_mapping = {
    'rad1': 'r_d=1',
    'rad2': 'r_d=2'
}

# Mapping for parameters to display names
param_display = {
    'beta_max': '\\beta_{max}',
    'encoder_conv2': 'd_Z',
    'weight_decay_cvae': '\\text{Weight Decay}',
    'lr_cvae': '\\text{Learning Rate}'
}

# Get list of all PDF files
pdf_files = [f for f in os.listdir(plots_dir) if f.endswith('.pdf')]

# Group files by dataset, rd, metric, and idx
file_groups = {}
for pdf in pdf_files:
    parts = pdf.replace('.pdf', '').split('_')
    
    # Extract components from filename
    # Format: dataset_rd_..._param_metric_idx#_ci_log.pdf
    dataset = None
    rd = None
    param = None
    metric = None
    idx = None
    
    # Find dataset
    if pdf.startswith('small_healthdgrid'):
        dataset = 'small_healthdgrid'
        remaining = pdf.replace('small_healthdgrid_', '')
    elif pdf.startswith('pm25grid'):
        dataset = 'pm25grid'
        remaining = pdf.replace('pm25grid_', '')
    if pdf.startswith('nonlocal_small_healthdgrid'):
        dataset = 'small_healthdgrid'
        remaining = pdf.replace('nonlocal_small_healthdgrid_', '')
    elif pdf.startswith('nonlocal_pm25grid'):
        dataset = 'pm25grid'
        remaining = pdf.replace('nonlocal_pm25grid_', '')
    else:
        continue
    
    # Find rd
    if 'rad1' in remaining:
        rd = 'rad1'
    elif 'rad2' in remaining:
        rd = 'rad2'
    else:
        continue
    
    # Find parameter
    for p in ['beta_max', 'encoder_conv2', 'weight_decay_cvae', 'lr_cvae']:
        if p in remaining:
            param = p
            break
    
    # Find metric
    if '_ate_' in remaining:
        metric = 'dir'
    elif '_spill_' in remaining:
        metric = 'spill'
    else:
        continue
    
    # Find idx
    if 'idx0' in remaining:
        idx = 0
    elif 'idx1' in remaining:
        idx = 1
    else:
        continue
    
    # Group by (dataset, rd, metric, idx)
    key = (dataset, rd, metric, idx)
    if key not in file_groups:
        file_groups[key] = []
    file_groups[key].append((param, pdf))

# Generate LaTeX code
latex_output = []

for (dataset, rd, metric, idx), files in sorted(file_groups.items()):
    if len(files) != 4:  # Should have 4 parameters
        continue
    
    # Sort files by parameter order
    param_order = ['beta_max', 'encoder_conv2', 'weight_decay_cvae', 'lr_cvae']
    files_sorted = sorted(files, key=lambda x: param_order.index(x[0]))
    
    # Get dataset info
    info = dataset_info.get(dataset, {})
    display_name = info.get('display_name', dataset)
    confounder = info.get(f'idx{idx}_confounder', 'Unknown')
    rd_display = rd_mapping.get(rd, rd)
    metric_upper = metric.lower()
    
    # Generate LaTeX figure environment
    latex_output.append("\\begin{figure}[tb]")
    latex_output.append("    \\centering")
    
    # Add subfigures for each parameter
    for i, (param, pdf_file) in enumerate(files_sorted):
        param_display_name = param_display.get(param, param)
        
        latex_output.append("    \\begin{subfigure}[b]{0.49\\textwidth}")
        latex_output.append("        \\centering")
        latex_output.append(f"        \\includegraphics[width=\\textwidth]{{{plots_dir}/{pdf_file}}}")
        # latex_output.append(f"        \\caption{{{param_display_name}}}")
        latex_output.append(f"        \\label{{fig:{dataset}_{rd}_{metric}_idx{idx}_{param}}}")
        latex_output.append("    \\end{subfigure}")
        
        # Add hfill between columns
        if i % 2 == 0:
            latex_output.append("    \\hfill")
        
        # Add line break after every 2 subfigures
        if i % 2 == 1 and i < len(files_sorted) - 1:
            latex_output.append("")
    
    # Add caption
    caption = (f"\\Rebuttal{{Sensitivity analysis for \\textsc{{C-VAE-unet}} models trained on spatial confounding environment $ {display_name} \\;({rd_display}) $ "
               f"with unobserved confounder ${confounder}$. "
               f"Each subplot shows \\textsc{{{metric_upper}}} as a function of a hyperparameter "
               f"across different neighborhood radii $r$. The error bounds represent the 95\\% confidence interval.}}")
    latex_output.append(f"    \\caption{{{caption}}}")
    latex_output.append(f"    \\label{{fig:nonlocal_{dataset}_{rd}_{metric}_idx{idx}}}")
    latex_output.append("    \\vspace{-1em}")
    latex_output.append("\\end{figure}")
    latex_output.append("")

# Write to file
with open(output_file, 'w') as f:
    f.write('\n'.join(latex_output))

print(f"Generated LaTeX code saved to: {output_file}")
print(f"Total figures generated: {len([k for k in file_groups.keys() if len(file_groups[k]) == 4])}")
```
